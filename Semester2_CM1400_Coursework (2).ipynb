**Question 2: Daily Mood Tracker**

**Real-Life Scenario:**

Mental wellbeing apps encourage users to track their daily mood to build self awareness and healthy habits. You have been asked to create a simple Mood Tracker that can be used by anyone who wants to record how they feel each day and reflect on their emotional patterns.

**Task:**

Create a Daily Mood Tracker program using classes and exception handling. The system should allow the user to:

•	Record their mood for the day (e.g., Happy, Sad, Stressed, Neutral).

•	Add an optional note describing what influenced their mood.

•	View all previous mood entries.

•	Ensure input is valid (e.g., prevent empty mood entries).

Your program must include:

•	try/except blocks to catch input errors.

•	Input validation to ensure users provide acceptable values.

•	Control flow using exception handling to manage invalid or unexpected user actions.


**Sample Output (for reference only):**

```python
# write your code here
class MoodTracker:
  def __init__(self):
    self.mood_entries = [] #creates a list for the moods and notes to be stored in

  def add_mood(self, mood, note):
    self.mood_entries.append({"mood": mood, "note": note}) #adds mood and note

  def view_moods(self):
    if len(self.mood_entries) == 0:  #checks if any moods entries stored
      print("No mood entries found")
    else:
      print("=== MOOD HISTORY ===")
      for i in range(len(self.mood_entries)): #loops through moods and notes then displays it
        mood = self.mood_entries[i]["mood"]
        note = self.mood_entries[i]["note"]
        print(f"{i+1}. {mood} | {note}")


tracker = MoodTracker()   #creates an instance of MoodTracker class

while True:    #main loop for program
  print("""=== Daily Mood Tracker ===
  1) Add Mood
  2) View Moods
  3) Exit
  """)

  try:
    choice = int(input("Choose (1-3): "))  #gets user input and coverts it to an INT

    if choice == 1: #adds new mood entry
      mood = input("How do you feel today? ").strip()    #strip() removes any lead or trailing white space

      if mood == "":
        raise ValueError("You must input a mood") #Validates string is not empty

      note = input("Add a short note (optional): ").strip()

      if note == "":   #adds a default note if no note added
        note = "No note added"
      tracker.add_mood(mood, note)
      print("Mood entry saved.")

    elif choice == 2:    #shows all saved moods and notes
      tracker.view_moods()

    elif choice == 3:
      print("Goodbye!")
      break

    else:    #handles invalid menu choices
      print("Invalid choice. Please choose from 1-3")

  except ValueError as e: #caputures invalid input errors like letters instead of numbers
    print("ERROR: ", e)

```

**Task Overview:**

You are provided with a CSV file containing housing data, 82 variables including price, area, Year.Built, Neighborhood, etc.. Your task is to write a Python program that performs targeted statistical analysis and generates multiple visualizations using the following libraries:

2.	statistics for computing descriptive statistics (Choose any 3 Numerical Variables)

Use the statistics module to compute any of the following for numerical features like Mean, Median, Mode, Standard Deviation, Minimum, Maximum and Range. You may also explore other statistical measures or apply them to additional features if relevant.

```python
# write your code here
import csv
import statistics
import matplotlib.pyplot as plt


#Create the lists to save the vaules
prices = []
areas = []
years = []

# open file
with open("ames.csv", "r", encoding="utf-8") as file:
  reader = csv.DictReader(file)

  # print first 5 rows to check im reading the file correctly
  # for i, row in enumerate(reader):
  #   print(row)
  #   if i == 4:
  #     break

  # get 3 parts of the full file
  # for i, row in enumerate(reader):
  #   print("price:", row["price"])
  #   print("Area:", row["area"])
  #   print("Year Built:", row["Year.Built"])
  #   print("--------------------")
  #   if i == 2:
  #     break

  # Collect the data
  for row in reader:
    try:
      if row["price"] != "" and row["area"] != "" and row["Year.Built"] != "":
        prices.append(float(row["price"]))
        areas.append(float(row["area"]))
        years.append(int(row["Year.Built"]))

    except ValueError:
      pass

# Check to see its worked
# print("Prices:", prices[:2])
# print("Areas:", areas[:2])
# print("Years:", years[:2])


#display the statistics
print("\n--- Price ---")
print("Mean:", statistics.mean(prices))
print("Median:", statistics.median(prices))
print("Minimum:", min(prices))
print("Maximum:", max(prices))
print("Range:", max(prices) - min(prices))

print("\n--- Area ---")
print("Mean:", statistics.mean(areas))
print("Median:", statistics.median(areas))
print("Minimum:", min(areas))
print("Maximum:", max(areas))
print("Range:", max(areas) - min(areas))

print("\n--- Year Built ---")
print("Mean:", statistics.mean(years))
print("Median:", statistics.median(years))
print("Minimum:", min(years))
print("Maximum:", max(years))
print("Range:", max(years) - min(years))


#plot history of prices
print("\n--- Price History ---")
plt.figure(figsize=(10, 6))
plt.hist(prices, bins=25, edgecolor="black")
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

#plot of areas
print("\n--- Area History ---")
plt.figure(figsize=(20, 25))
plt.scatter(areas, prices, alpha=0.5)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()


# Define the bins for the area ranges.
bins = [0, 250, 500, 750, 1000, 1250, 1500, 1750, 2000, 2250,
        2500, 2750, 3000, 3250, 3500, 3750, 4000, 4250, 4500,
        4750, 5000, 5250, 5500, 5750, 6000, 6250, 6500, 6750]

# create empty bins
grouped_prices= {i: [] for i in range(len(bins) - 1)}

# assign each house to a bin
for i in range(len(areas)):
  area = areas[i]
  price = prices[i]

  for j in range(len(bins) - 1):
    if bins[j] <= area < bins[j + 1]:
      grouped_prices[j].append(price)
      break

average_price = []
mid_points = []

for j in grouped_prices:
  if len(grouped_prices[j]) > 0:
    avg = sum(grouped_prices[j]) / len(grouped_prices[j])
    average_price.append(avg)

    # X axis
    mid = (bins[j + 1] + bins[j]) / 2
    mid_points.append(mid)


#line graph
print("\n--- Average Price by Area Range ---")
plt.figure(figsize=(10, 6))
plt.plot(mid_points, average_price, marker="o")
plt.title("Average Price by Area Range")
plt.xlabel("Area")
plt.ylabel("Average Price")
plt.show()
```
